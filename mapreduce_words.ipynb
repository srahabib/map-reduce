# Partitioning Phase

```python
string = 'welcome welcome welcome to to the the the the the world of of of of python python python programming'
string = string.split(' ')
li = []
for i in range(len(string)):
    li.append([string[i]])
    
print(li)
```

# Mapping Phase

```python
def mapper(string):
    mapped = {}
    for i in string :
        mapped[i] = 1
        
    return(mapped)

m = list(map(mapper,li))
m
```
